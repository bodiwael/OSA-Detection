```python
# creating reference to root node
Age = db.reference("Data/Age")

Weight = db.reference("Data/Weight")

Height = db.reference("Data/Height")

Gender = db.reference("Data/Gender")

SPO2 = db.reference("Data/SPO2")

HR = db.reference("Data/HR")
```

```python
# retrieving data from root node
age = Age.get()
weight = Weight.get()
height = Height.get()
gender = Gender.get()
spo2 = SPO2.get()
hr = HR.get()

print(gender)
```

```python
BMI = int(weight) / (int(height) ** 2)
```

```python
import pandas as pd
import joblib

model_path = "/kaggle/input/random-forest-model/rf_model.pkl"

loaded_model = joblib.load(model_path)

gender_mapping = {"male": 0, "female": 1}
gender_num = gender_mapping.get(gender.lower(), -1)
        
new_data = {'HR': hr, 'SpO2': spo2, 'Age': age, 'Gender': gender_num, 'BMI': BMI}

# Convert the data to a feature vector
feature_vector = pd.DataFrame([new_data])

rf_predictions = loaded_model.predict(feature_vector)

print(rf_predictions[0])
```
